```python
import requests
from bs4 import BeautifulSoup
import time
```

```python
r = requests.get('https://app.tzuchi.com.tw/tchw/opdreg/opdstatus/loginstatus.aspx?DShospital=TCWSopd')
rs = BeautifulSoup(r.text, 'html.parser')
rs
```

```python
all_list = {}
for i in rs.find('table').findAll('tr')[3:]:
    if i.findAll('td')[0].text[-1] == '科':
        part = str(i.findAll('td')[0].text)
        if (part not in all_list):
            all_list[part] = []
        all_list[part].append({
            'url': i.findAll('td')[2].find('a')['href']
        })
all_list
#     break
```

```python
now = time.time()
crawl_data = []
for i in all_list['中醫科']:
    r = requests.get(i['url'])
    print('開始爬取'+ i['url'])
    rs = BeautifulSoup(r.text, 'html.parser')
    doctor_num = rs.find('table').findAll('td')[5].text
    doctor = rs.find('table').findAll('td')[7].text
    num = rs.find('table').findAll('td')[13].text
    next_num = rs.find('table').findAll('td')[15].text
    fini_num = rs.find('table').findAll('td')[17].text
    crawl_data.append({
        'doctor_num': doctor_num,
        'doctor': doctor,
        'num': num,
        'fini_num': fini_num,
        'next_num': next_num,
    })
print(time.time() - now)
crawl_data
```
